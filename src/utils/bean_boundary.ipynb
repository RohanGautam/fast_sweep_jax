```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = "../../data/beans.png"
image = Image.open(img)
image
```

```python
greyscale_image = image.convert("L")
greyscale_image
```

```python
image_array = np.array(greyscale_image)
image_array.shape
```

```python
image_array[image_array > 100] = 255
image_array[image_array <= 100] = 0
padded_image = np.pad(image_array, pad_width=10, mode="constant", constant_values=255)
```

```python
plt.imshow(padded_image, cmap="grey")
```

```python
output_image = Image.fromarray(padded_image)
output_image.save("../../data/processed_beans.png")
```
